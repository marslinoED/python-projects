## matplotlib Tutorial

###### import library

```python
import matplotlib.pyplot as plt
```

###### making two lists

```python
# x axis
x=[1,2,3,4,5,6,7]

# y axis 
y=[50,51,52,48,47,49,46]
```

###### xlabel, ylabel, title, plot, show

```python
# x axis name
plt.xlabel('Day')

# y axis name
plt.ylabel('Temperature')

# pic name
plt.title('Weather')

# grid
#plt.grid(True)

# remove axis
#plt.axis('off')

# making the plot, (x,y, color of plot('color name' or 'hexa decimal color') , linewitdth, linestyle)
# line style '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted' 
plt.plot(x,y,color='#d6ac38',linewidth=3, linestyle = "-.")

# print the plot
plt.show()
```

###### More Formats

```python
#format string [color][marker] [line style]
plt.plot(x,y,'ro--') 

plt.show()
```

```python
# Same effect as 'b+--' format string
plt.plot(x,y,color='blue',marker='+',linestyle='--',markersize=10)

plt.show()
```

<h6>markersize</h6>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)

plt.show()
```

###### alpha property to control transparency of line chart

```python
# alpha can be specified on a scale 0 to 1
plt.plot(x,y,'bo',markersize = "12",alpha=0.4) 

plt.show()
```

###### Axes Labels, Legend

```python
days=[1,2,3,4,5,6,7]
max_t=[47,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,42]
```

<h3>Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGLE','AMAZONE','MSFT','FB']
revenue=[90,136,85,27]
```

```python
plt.bar(company,revenue, label="Revenue")
plt.xlabel("Companies")
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
import numpy as np
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.4,profit, width=0.4,label="Profit")
# changing the name
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
# printing on each other
plt.bar(xpos,revenue, width=0.4, label="Revenue")
plt.bar(xpos,profit, width=0.4,label="Profit")
# changing the name
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3>Horizontal bar chart using barh function<h3>

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
plt.show()
```

#### histogram

```python
# relation between sugar and count of sugar
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.3) # by default number of bins is set to 10
plt.show()
```

### bins parameter</h3>

```python
plt.hist(blood_sugar,rwidth=0.5,bins=3)
plt.show()
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=20, rwidth=0.95, color='g')
```

### Mutiple data samples in a histogram</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[77,100,125,150], rwidth=0.8, color=['grey','orange'],label=['men','women'])
plt.legend()
plt.show()
```

##### histtype

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")
# step 'bar', 'barstacked', 'step', 'stepfilled'
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
plt.show()
```

### horizontal orientation</h3>

```python
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
plt.show()
```

### Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,300,250] #expenses 
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

### Shadow</h3>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

### Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',radius=1.66)
plt.show()
```

### Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.4f%%',radius=1.5,explode=[.2,0,0,0,0])
plt.show() # formatting 
```

### counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=180)
plt.show()
```

<h2 align="center">Saving chart to a file using savefig</h2>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=180)
plt.savefig("Figuers/piechart.png", bbox_inches="tight", pad_inches=7, transparent=True)
```
