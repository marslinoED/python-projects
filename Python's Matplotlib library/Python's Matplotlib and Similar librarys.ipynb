## matplotlib Tutorial

###### import library

```python
import matplotlib.pyplot as plt
import numpy as np
```

###### making two lists

```python
# x axis
x=[1,2,3,4,5,6,7]

# y axis 
y=[50,51,52,48,47,49,46]
```

###### xlabel, ylabel, xticks, yticks, title, plot, show

```python
# x axis name
plt.xlabel('Day')

# y axis name
plt.ylabel('Temperature')

# pic name
# fontdict to change font name and size
plt.title('Weather', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 18})

# grid
#plt.grid(True)

# remove axis
#plt.axis('off')

# making the plot, (x,y, color of plot('color name' or 'hexa decimal color') , linewitdth, linestyle)
# line style '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted' 
plt.plot(x,y,color='#d6ac38',linewidth=3, linestyle = "-.")

plt.xticks(np.arange(0,8))
plt.yticks(np.arange(46,53))

# print the plot
plt.show()
```

###### More Formats

```python
#format string [color][marker] [line style]
plt.plot(x,y,'ro--') 

plt.show()
```

```python
# Same effect as 'b+--' format string
plt.plot(x,y,color='blue',marker='+',linestyle='--',markersize=10)

plt.show()
```

<h6>markersize</h6>

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)

plt.show()
```

###### alpha property to control transparency of line chart

```python
# alpha can be specified on a scale 0 to 1
plt.plot(x,y,'bo',markersize = "12",alpha=0.4) 

plt.show()
```

### Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,300,250] #expenses 
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

### Shadow</h3>

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

### Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.3f%%',radius=1.66)
plt.show()
```

### Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.4f%%',radius=1.5,explode=[.2,0,0,0,0])
plt.show() # formatting 
```

### counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=180)
plt.show()
```

<h1 align = "center" > Animation

```python

import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML
# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)
anim = animation.FuncAnimation(fig, animate, init_func=init,frames=100, interval=20, blit=True)
rc('animation', html='html5')
anim
```
