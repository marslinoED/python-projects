```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model('model/digits.h5')

def prediction(theimg):
    img = cv2.cvtColor(theimg, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(28, 28))
    img = img / 255
    img = img.reshape(1,28,28,1)
    result = np.argmax( model.predict(img)) 
    return result
```

```python

cap = cv2.VideoCapture(0 + cv2.CAP_DSHOW)
WIDTH = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
HEIGHT = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

while True:

    _, frame = cap.read()

    bbox_size = (60, 60)
    bbox = [(int(WIDTH // 2 - bbox_size[0] // 2), int(HEIGHT // 2 - bbox_size[1] // 2)),
            (int(WIDTH // 2 + bbox_size[0] // 2), int(HEIGHT // 2 + bbox_size[1] // 2))]
    
    cv2.rectangle(frame,bbox[0],bbox[1],(250,255,255),2)
    img_cropped = frame[bbox[0][1]:bbox[1][1], bbox[0][0]:bbox[1][0]]
    img_gray = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2GRAY)
    img_gray = cv2.resize(img_gray, (200, 200))
    cv2.imshow("cropped", img_gray)
    
    
    print(prediction(img_cropped))
    cv2.putText(frame, "Result: " + str(prediction(img_cropped)), (100,100), cv2.FONT_HERSHEY_COMPLEX,0.8,(255,0,255),2)
  

    cv2.imshow("input", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

model = load_model('model/digits.h5')

def preprocess_image(theimg):
    img = cv2.cvtColor(theimg, cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(28, 28))
    img = img / 255
    img = img.reshape(1,28,28,1)
    return img

def predict_batch(images):
    preprocessed_images = np.array([preprocess_image(img) for img in images])
    predictions = np.argmax(model.predict(preprocessed_images), axis=1)
    return predictions

paths = []
for i in range(10):
    for j in range(10):
        paths.append("Cells/cell["+str(i)+"]["+str(j)+"].jpg")

batch_size = 8
for i in range(20, 100, batch_size):
    batch_paths = paths[i:i+batch_size]
    batch_images = [cv2.imread(path) for path in batch_paths]
    predictions = predict_batch(batch_images)
    print(predictions)
```
