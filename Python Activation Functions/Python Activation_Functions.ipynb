```python
import numpy as np
import matplotlib.pyplot as plt
```

## Activation Functions:

##### Sigmoid Activation Function:
Use Case:<br>
Binary Classification Problems<br>
Pros: Well-suited for binary classification tasks where the goal is to output probabilities; smooth gradient across most of its range.<br>
Cons: Prone to the vanishing gradient problem, where gradients become extremely small during backpropagation.<br>

##### Tanh Activation Function:
Use Case:<br>
Similar to Sigmoid, but centered around zero<br>
Pros: Captures negative and positive values; suitable for tasks where the output needs to be centered around zero.<br>
Cons: Still susceptible to the vanishing gradient problem.<br>

##### ReLU (Rectified Linear Unit) Activation Function:
Use Case:<br>
General-purpose, hidden layers in most cases<br>
Pros: Efficient training due to simplicity; alleviates vanishing gradient problem; computationally efficient.<br>
Cons: Can suffer from the "dying ReLU" problem where neurons can become inactive during training.<br>
##### Leaky ReLU Activation Function:

Use Case: Address the "dying ReLU" problem<br>
Pros: Allows a small, non-zero gradient for negative inputs, preventing neurons from becoming completely inactive.<br>
Cons: May not perform well on all tasks; sensitivity to the choice of the leaky parameter.<br>

##### Other Activation Functions (e.g., Swish, PReLU, ELU):
Use Case: Experimentation and task-specific tuning<br>
Pros: Address specific issues or introduce non-linearities in different ways.<br>
Cons: May have computational overhead or require careful tuning; not universally applicable.<br>
##### Choosing an Activation Function:

For hidden layers in a neural network, ReLU and its variants like Leaky ReLU are commonly used due to their efficiency and ability to mitigate the vanishing gradient problem.<br>
For the output layer in binary classification tasks, Sigmoid is often suitable.<br>
For regression tasks or multi-class classification, Softmax is commonly used in the output layer.<br>
It's important to experiment with different activation functions and observe their impact on the model's performance.<br> The choice may also be influenced by the architecture of the neural network and specific characteristics of the data.

## Math:

<h3 style=color:darkblue> 1. Sigmoid Activation Function:</h3>
<h5 style=color:darkred>  Definition:</h5>
Y(x) = 1+e−x1
<h5 style=color:darkred> Characteristics:</h5>
• Output range: (0, 1)<br>
• S-shaped curve<br>
• Used in binary classification problems


```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

<h3 style=color:darkblue> 2. Tanh Activation Function:</h3>
<h5 style=color:darkred>  Definition:</h5>
Y(x) = (ex+e−x)/(ex−e−x) = Tanh(x)
<h5 style=color:darkred> Characteristics:</h5>
• Output range: (0, 1)<br>
• Similar to Sigmoid but with a symmetric range


```python
def tanh(x):
    return np.tanh(x)
```

<h3 style=color:darkblue>  3. ReLU (Rectified Linear Unit) Activation Function:</h3>
<h5 style=color:darkred>  Definition:</h5>
Y(x) = Max(0,x)
<h5 style=color:darkred> Characteristics:</h5>
• Output is zero for negative inputs, linear for positive inputs<br>
• Simple and computationally efficient<br>
• Prone to the "dying ReLU" problem for negative inputs


```python
def relu(x):
    return np.maximum(0, x)
```

<h3 style=color:darkblue>  4. Leaky ReLU Activation Function:</h3>
<h5 style=color:darkred>  Definition:</h5>
Y(x) =
<pre>
x  if x >  0
ax if x <= 0
</pre>
<h5 style=color:darkred> Characteristics:</h5>
• Introduces a small slope for negative inputs, addressing the "dying ReLU" problem,<br>
• Can prevent neurons from becoming inactive

```python
def leaky_relu(x, alpha=0.1):
    return np.maximum(alpha * x, x)
```

<h3 style=color:darkblue>  5. SoftMax:</h3>
<h5 style=color:darkred> Characteristics:</h5>
• Multi-class Classification Problems:<br>
Softmax is commonly used in the output layer of neural networks for tasks where the input can belong to one of multiple classes.

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtracting max(x) for numerical stability
    return exp_x / np.sum(exp_x, axis=0)
```

#### Visualization:

```python
# Visualization
x_values = np.linspace(-10, 10, 100)
plt.figure(figsize=(10, 11))

# Sigmoid
plt.subplot(3, 2, 1)
plt.plot(x_values, sigmoid(x_values))
plt.title('Sigmoid Activation Function')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# Tanh
plt.subplot(3, 2, 2)
plt.plot(x_values, tanh(x_values))
plt.title('Tanh Activation Function')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# ReLU
plt.subplot(3, 2, 3)
plt.plot(x_values, relu(x_values))
plt.title('ReLU Activation Function')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# Leaky ReLU
plt.subplot(3, 2, 4)
plt.plot(x_values, leaky_relu(x_values))
plt.title('Leaky ReLU Activation Function')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

# Softmax
plt.subplot(3, 2, 5)
plt.plot(x_values, softmax(x_values))
plt.title('Softmax')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)

plt.subplots_adjust(hspace=0.2, wspace=0.2)
plt.show()
```
