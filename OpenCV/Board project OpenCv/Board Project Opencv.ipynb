```python
import cv2
import mediapipe as mp
import math
import numpy as np

i = [0] * 100
# Create a blank canvas
canvas = np.zeros((480, 640, 3), dtype=np.uint8)

def CheckIfClosser(x,y,tx,ty,lx,ly):
    if ((tx <= x <= lx) and (ty <= y <= ly)):
        return (True)
    
    
def DrawrTheRectangle(tx,ty,lx,ly,Which):

    i[Which] += 4
    cv2.rectangle(img, (tx,ty), (tx + i[Which], ly), (0,255, 0), thickness=cv2.FILLED)
    if (tx + i[Which] >= lx):
        i[Which] = lx - tx - 4
        return (True)

def GetDimensions(x, y, width, height):
    return int(x), int(y), int(width + x), int(height + y)

def GetAxes(landmarks, FingerTip):
        # Get the coordinates of landmarks
        x, y = landmarks[FingerTip].x, landmarks[FingerTip].y
        # Draw circles at the landmark positions
        h, w, c = img.shape
        cx, cy = int(x * w), int(y * h)
        return int(cx), int(cy)
    
def DrawFinger(cx,cy,COLOR):  
        # this part to draw the hand body
        cv2.circle(img, (cx, cy), 5, COLOR, thickness=cv2.FILLED)
def Distance(x1,y1,x2,y2):
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

def SIZEBUTTON(z1,z2,z3,z4, Which,size):
    
        # Getting the dimensins
        x,y,width,height = GetDimensions(z1,z2,z3,z4)
        
        # Drawing The Button
        cv2.rectangle(img, (x,y), (width,height), (139,0, 0), thickness=cv2.FILLED)
        cv2.putText(img, str(size), (x+5,int((y+height)/2)), cv2.FONT_HERSHEY_SIMPLEX,0.5 * ((width - x)/40) , (255, 255, 255), 1)
        
        
        if(CheckIfClosser(cx12, cy12, x,y,width,height)):
              if(DrawrTheRectangle(x,height+5,width,height+10,Which)):
                    # cv2.putText(img, "CHECK", (x+5,int((y+height)/2)), cv2.FONT_HERSHEY_SIMPLEX,0.3 * ((width - x)/40) , (255, 255, 255), 1)
                    cv2.rectangle(img, (x,y), (width,height), (255,255, 255), thickness= int(1 * ((width - x)/40)) )
                    global helloKetti
                    helloKetti = size
                    
                    
                    
        else:
            i[Which] = 0
        
        # Drawing The Box
        cv2.rectangle(img, (x, height+5), (width, height+10), (128,128,128), thickness=1)
        
def ERASEBUTTON(z1,z2,z3,z4, Which):
    
        # Getting the dimensins
        x,y,width,height = GetDimensions(z1,z2,z3,z4)
        
        # Drawing The Button
        cv2.putText(img, "ERASE", (x+5-10,int((y+height)/2)), cv2.FONT_HERSHEY_SIMPLEX,0.7 * ((width - x)/40) , (255, 255, 255), 1)
        cv2.putText(img, " ALL", (x+5-10,int((y+height)/2 + 10)), cv2.FONT_HERSHEY_SIMPLEX,0.7 * ((width - x)/40) , (255, 255, 255), 1)
        
        
        if(CheckIfClosser(cx12, cy12, x,y,width,height)):
              if(DrawrTheRectangle(x,height+5,width,height+10,Which)):
                    # cv2.putText(img, "CHECK", (x+5,int((y+height)/2)), cv2.FONT_HERSHEY_SIMPLEX,0.3 * ((width - x)/40) , (255, 255, 255), 1)
                    cv2.rectangle(img, (x,y), (width,height), (255,255, 255), thickness= int(1 * ((width - x)/40)) )
                    global canvas
                    canvas = np.zeros((480, 640, 3), dtype=np.uint8)
                    
                    
                    
        else:
            i[Which] = 0
        
        # Drawing The Box
        cv2.rectangle(img, (x, height+5), (width, height+10), (128,128,128), thickness=1)


def COLORBUTTON(z1,z2,z3,z4, Which,COLOR):
    
        # Getting the dimensins
        x,y,width,height = GetDimensions(z1,z2,z3,z4)
        
        # Drawing The Button
        cv2.rectangle(img, (x,y), (width,height),COLOR, thickness=cv2.FILLED)
        
        
        if(CheckIfClosser(cx12, cy12, x,y,width,height)):
              if(DrawrTheRectangle(x,height+5,width,height+10,Which)):
                    # cv2.putText(img, "CHECK", (x+5,int((y+height)/2)), cv2.FONT_HERSHEY_SIMPLEX,0.3 * ((width - x)/40) , (255, 255, 255), 1)
                    cv2.rectangle(img, (x,y), (width,height), (255,255, 255), thickness= int(1 * ((width - x)/40)) )
                    global COLORis
                    COLORis = COLOR
                    
                    
                    
        else:
            i[Which] = 0
        
        # Drawing The Box
        cv2.rectangle(img, (x, height+5), (width, height+10), (128,128,128), thickness=1)

def draw_line(image, start_point, end_point, pen_size,COLOR):
    cv2.line(image, start_point, end_point, COLOR, pen_size)

# Function to change color of pixels touched by finger tip
def change_color(image, x, y, pen_size,COLOR):
    global prev_point
    
    # Change color of the pixel at (x, y)
    image[y, x] = (COLOR)  # White color
    
    # Draw line from previous point to current point
    if prev_point is not None:
        draw_line(image, prev_point, (x, y),pen_size,COLOR)
        
    # Update previous point
    prev_point = (x, y)
    
def erase_pixels(image, x, y, erase_range):
    image[max(0, y - erase_range):min(image.shape[0], y + erase_range), 
          max(0, x - erase_range):min(image.shape[1], x + erase_range)] = 0
```

```python
cap = cv2.VideoCapture(0)

mpHands = mp.solutions.hands
hands = mpHands.Hands()
mpDraw = mp.solutions.drawing_utils

# Create a blank canvas
canvas = np.zeros((480, 640, 3), dtype=np.uint8)
# Initialize variables to track previous point
prev_point = None
helloKetti = 3
COLORis = (0,255,0)

while True:
    success, img = cap.read()
    img = cv2.flip(img,1)
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    results = hands.process(imgRGB)
    
    # Getting the land mark
    if results.multi_hand_landmarks:
        for handLms in results.multi_hand_landmarks:
            for id, lm in enumerate(handLms.landmark):
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
        landmarks = handLms.landmark
                
        # Drawing the index finger
        cx8,cy8   = GetAxes(landmarks,8)
        cx12,cy12 = GetAxes(landmarks, 12)
        cx16,cy16 = GetAxes(landmarks, 16)
       
        if (Distance(cx8,cy8,cx12,cy12) > 40):
            DrawFinger(cx8,cy8,(255,255,255))
            change_color(canvas, cx8, cy8,helloKetti, COLORis)
            
        elif (Distance(cx12,cy12,cx16,cy16) > 40):
            DrawFinger(cx12,cy12,(255,255,255))
            erase_pixels(canvas,cx12,cy12,erase_range=20)
            prev_point = None
        else:
            DrawFinger(cx16,cy16,(255,255,255))
            prev_point = None
        
        SIZEBUTTON(600,50,20,20,0,1)
        SIZEBUTTON(600,90,20,20,1,3)
        SIZEBUTTON(600,130,20,20,2,5)
        COLORBUTTON(600,170,20,20,3,(0,0,255))
        COLORBUTTON(600,210,20,20,4,(255,0,0))
        COLORBUTTON(600,250,20,20,5,(0,255,0))
        ERASEBUTTON(600,290,20,20,6)
        
        

    # Display canvas
    # cv2.imshow('Canvas', cv2.add(canvas,img))
     # Create a mask for the canvas
        
    mask = cv2.cvtColor(canvas, cv2.COLOR_BGR2GRAY)
    _, mask = cv2.threshold(mask, 1, 255, cv2.THRESH_BINARY)

    # Invert the mask
    mask_inv = cv2.bitwise_not(mask)

    # Apply the mask to the canvas
    canvas_masked = cv2.bitwise_and(canvas, canvas, mask=mask)

    # Apply the inverted mask to the frame
    frame_masked = cv2.bitwise_and(img, img, mask=mask_inv)

    # Combine the masked canvas and frame
    frame_with_canvas = cv2.add(canvas_masked, frame_masked)

    # Display frame with canvas
    cv2.imshow('Frame', frame_with_canvas)
    
    # Check for the 'Esc' key press
    key = cv2.waitKey(1)
    if key == 27:  # 27 is the keycode for 'Esc'
        break

cap.release()
cv2.destroyAllWindows()
```
